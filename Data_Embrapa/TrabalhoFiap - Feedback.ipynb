##Produção/Comercialização


```python
listaUrl = ['Producao','Comercio']
for x in listaUrl:
  urlsProducaoComercializacao = f'http://vitibrasil.cnpuv.embrapa.br/download/{x}.csv'
  dados = pd.read_csv(urlsProducaoComercializacao, sep=';')
  dados = dados.iloc[:, 2:]
  if x == 'Producao':
    indices_maiusculas = dados.query('produto.str.isupper()', engine='python').index
  else:
    indices_maiusculas = dados.query('Produto.str.isupper()', engine='python').index      # Aqui tem uma diferença na variável produto. 

  dados['categoria'] = ''                                    # Sempre tomar cuidado com a atribução de dados á colunas. É mais fácil atribuir os dados prontos á uma nova coluna do que declarar uma coluna vazia

  categoria_atual = None
  for idx in indices_maiusculas:
      if x == 'Producao':
        categoria_atual = dados.loc[idx, 'produto']
      else:
         categoria_atual = dados.loc[idx, 'Produto']         # Você sempre precisa ter cuidado com os erros de formatação. São simples mas as vezes são chatos de achar
      dados.loc[idx + 1:, 'categoria'] = categoria_atual

  dados.drop(indices_maiusculas, axis=0, inplace=True)
  coluna_movida = dados.pop('categoria')                     # Talvez houvesse um jeito melhor de selecionar esses dados
  dados.insert(1, 'categoria', coluna_movida)                # Alguns operardores de DataFrames podem seguir a lógica de outras estruturas de dados
  dados = dados.replace('*', '0')
  dados = dados.replace('nd', '0')
  dados = dados.fillna(0)
  colunas_para_converter = dados.columns[2:]
  dados[colunas_para_converter] = dados[colunas_para_converter].astype('int64')
  dados_PC = dados.replace(r'^\s*|\s*$', '', regex=True)     # Evite usar o mesmo nome de variável (dados) para todos os estados do dataset. Experimente utilizar um nome por etapa da transformação
  dados_PC.to_json(f'{x}.json')                              # Evite usar o mesmo nome de variável para todos os estados do dataset. Experimente utilizar um nome por etapa da transformação
#dados
```

##Importação/Exportação

```python
listaUrl = ['ExpEspumantes','ExpSuco','ExpUva','ExpVinho','ImpEspumantes','ImpFrescas','ImpPassas','ImpSuco','ImpVinhos']
for x in listaUrl:
  urlsExpImp = f'http://vitibrasil.cnpuv.embrapa.br/download/{x}.csv'

  dados = pd.read_csv(urlsExpImp, sep=';')
  dados = dados.iloc[:, 1:]

  dados = dados.replace('*', '0')
  dados = dados.replace('nd', '0')
  dados_IE = dados.fillna(0)
  colunas_para_converter = dados_IE.columns[1:]
  dados[colunas_para_converter] = dados_IE[colunas_para_converter].astype('int64')
  dados_IE.to_json(f'{x}.json')
#dados
```

```python
dados_IE
```

##Processamento

```python
listaUrl = ['ProcessaAmericanas','ProcessaMesa','ProcessaSemclass','ProcessaViniferas']
for x in listaUrl:
  urlProcessamento = f'http://vitibrasil.cnpuv.embrapa.br/download/{x}.csv'
  dados = pd.read_csv(urlProcessamento, sep='\t')
  dados = dados.iloc[:, 2:]

  indices_maiusculas = dados.query('cultivar.str.isupper()', engine='python').index

  dados['categoria'] = ''

  categoria_atual = None
  for idx in indices_maiusculas:
      categoria_atual = dados.loc[idx, 'cultivar']
      dados.loc[idx + 1:, 'categoria'] = categoria_atual

  dados.drop(indices_maiusculas, axis=0, inplace=True)
  coluna_movida = dados.pop('categoria')  # Remove a coluna e a armazena em 'coluna_movida'
  dados.insert(1, 'categoria', coluna_movida)  # Insere a coluna na segunda posição (índice 1)
  dados = dados.replace('*', '0')
  dados = dados.replace('nd', '0')
  dados_PR = dados.fillna(0)
  colunas_para_converter = dados_PR.columns[2:]  # Selecionando as colunas a partir da segunda
  dados_PR[colunas_para_converter] = dados_PR[colunas_para_converter].astype('int64')
  dados_PR.to_json(f'{x}.json')
#dados
```
